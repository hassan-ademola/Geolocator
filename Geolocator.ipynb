```python
import urllib
import json
import tkinter as tk

window = tk.Tk()
window.geometry('600x500')
window.title('Geolocator')

bgimage = tk.PhotoImage(file ='images/landscape.png')
bglabel = tk.Label(window, image = bgimage)
bglabel.place(relwidth = 1, relheight = 1)

def fetched():
    api_key = 42
    serviceurl = 'http://py4e-data.dr-chuck.net/json?'
    address = entry.get()
    if len(address) < 1: 
        result['text'] = 'Please, enter a valid address'
        result['fg'] = 'red'
        return 
    parms = dict()
    parms['address'] = address
    parms['key'] = api_key
    url = serviceurl + urllib.parse.urlencode(parms)
    
    try:
        uh = urllib.request.urlopen(url)
        data = uh.read().decode()
        js = json.loads(data)
        lat = js['results'][0]['geometry']['location']['lat']
        lng = js['results'][0]['geometry']['location']['lng']
        location = js['results'][0]['formatted_address']
        display = 'Latitude: {0}\nLongitude: {1}\nLocation: {2}'.format(lat, lng, location)
        result['fg'] = 'black'
        result['text'] = display
        
    except urllib.error.URLError:
        result['fg'] = 'red'
        result['text'] = 'Please, check your internet connection'
    except:
        result['text'] =  'Failure To Retrieve'
        result['fg'] = 'red'
                
def quit():
    window.destroy()
        


frame = tk.Frame(window, bg = '#8dc2fe', bd = 2)
frame.place(relheight = 0.1, relwidth = 0.9, rely = 0.1, relx = 0.5, anchor = 'n')

title = tk.Label(frame, text = 'Enter Address', font=('TkDefaultFont', 11), bg = '#dff0d4')
title.place(relheight = 1, relwidth = 0.207)

entry = tk.Entry(frame, bd = 0, font = ('TkDefaultFont', 14))
entry.place(relheight = 1, relwidth = 0.58, relx = 0.21)
entry.focus_set()

fetch = tk.Button(frame, text = 'Fetch', font=('TkDefaultFont', 11), bg = '#dff0d4', command = fetched)
fetch.place(relheight = 1, relwidth = 0.207, relx = 0.793)

lframe = tk.Frame(window, bg = '#8dc2fe', bd = 2)
lframe.place(relheight = 0.6, relwidth = 0.9, rely = 0.25, relx = 0.5, anchor = 'n')


result = tk.Label(lframe, bg = 'white', font=('TkDefaultFont', 14), anchor = 'nw', justify = 'left', wraplength=500)
result.place(relheight = 1, relwidth = 1)

llframe= tk.Frame(window, bg = '#8dc2fe', bd = 2)
llframe.place(relheight = 0.1 , relwidth = 0.25, relx = 0.5, anchor = 'n', rely = 0.87)

quit = tk.Button(llframe, text = 'Quit', font=('TkDefaultFont', 11), bg = '#dff0d4', command = quit)
quit.place(relheight = 1, relwidth = 1)

window.mainloop()
```
